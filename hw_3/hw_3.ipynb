```python
#3 процент выживаемости у каждого класса пассажиров
```

```python
passengers = dict(df['Pclass'].value_counts())
passengers_survived = dict(df.groupby("Pclass")["Survived"].sum())
for i in range(1,4):
    procent = round(passengers_survived[i]/passengers[i]*100, 2)
    print(f'Процент выживаемости у {i} класса составляет {procent}%')
```

```python
#4.2  Самое популярное женское имя
```

```python
df_Women = df[df["Sex"]=='female']
df_Women
```

```python
df_Women = df[df["Sex"]=='female']
name_List = list(df_Women["Name"])
name_Mrs = []
name_Miss = []
name_Women = []
for name in name_List:
    if "Mrs" in name:
        name_Mrs.append(name)
    else:
        name_Miss.append(name)
for i in range(0, len(name_Miss)):
    names = name_Miss[i].split(". ")
    names = names[1].split()
    name_Women.append(names[0])
for name in name_Mrs:
    names = name.split()
    for i in range(0, len(names)):
        if '(' in names[i]:
            nameW = names[i].lstrip("(")
            nameW = nameW.rstrip(")")
    name_Women.append(nameW)
    
print(max(set(name_Women), key = name_Women.count))
```

```python
#5 Выведите самое популярное мужское и самое популярное женское имя на корабле в каждом классе
```

```python
#5.1 Самое популярное мужское имя в разных классах
```

```python
for i in range(1,3):
    df_Men = df[(df["Pclass"]==i) & (df["Sex"] == 'male')]
    df_Men = df_Men["Name"].str.split('.',expand=True)
    df_Men.columns=["Surname", "Name"]
    df_Men = df_Men["Name"].str.split(' ',expand=True)
    df_Men.columns = ["1", "Name", "Second Name", "4", "5", "6"]
    print(f'Самое популярное имя в классе {i}: {df_Men["Name"].value_counts().index[0]}')
df_Men = df[(df["Pclass"]==3) & (df["Sex"] == 'male')]
df_Men = df_Men["Name"].str.split('.',expand=True)
df_Men.columns=["Surname", "Name"]
df_Men = df_Men["Name"].str.split(' ',expand=True)
df_Men.columns = ["1", "Name", "Second Name", "4", "5"]
print(f'Самое популярное имя в классе 3: {df_Men["Name"].value_counts().index[0]}')
```

```python
#5.2 Самое популярное женское имя в разных классах
```

```python
for cl in range(1, 4):
    df_Women = df[(df["Sex"]=='female') & (df["Pclass"]== cl)]
    name_List = list(df_Women["Name"])
    name_Mrs = []
    name_Miss = []
    name_Women = []
    for name in name_List:
        if "Mrs" in name:
            name_Mrs.append(name)
        else:
            name_Miss.append(name)
    for i in range(0, len(name_Miss)):
        names = name_Miss[i].split(". ")
        names = names[1].split()
        name_Women.append(names[0])
    for name in name_Mrs:
        names = name.split()
        for i in range(0, len(names)):
            if '(' in names[i]:
                nameW = names[i].lstrip("(")
                nameW = nameW.rstrip(")")
        name_Women.append(nameW)
    popName = max(set(name_Women), key = name_Women.count)
    print(f'Самое популярное имя в классе {cl}: {popName}')
```

```python
#8 Выведите количества n-местных кабин (кают?) , в которых было 2, 3, 4, .... человек
```

```python
cabins = list(df['Cabin'].dropna())

uniq = set(cabins)
cabinsCount = []
for i in uniq:
    passengers = cabins.count(i)
    cabinsCount.append(passengers)
passQuant = set(cabinsCount)
for i in passQuant:
    n = cabinsCount.count(i)
    print(f'{i}-местных кабин было {n}')
```
