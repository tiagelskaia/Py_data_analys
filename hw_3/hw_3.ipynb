```python
#3 процент выживаемости у каждого класса пассажиров
```

```python
passengers = dict(df['Pclass'].value_counts())
passengers_survived = dict(df.groupby("Pclass")["Survived"].sum())
for i in range(1,4):
    procent = round(passengers_survived[i]/passengers[i]*100, 2)
    print(f'Процент выживаемости у {i} класса составляет {procent}%')
```
